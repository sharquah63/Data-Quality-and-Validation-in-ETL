Q1. Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

-> Data quality in ETL pipelines refers to the accuracy, completeness, consistency, timeliness, and validity of data as it moves through Extract, Transform, and Load stages. It exceeds basic cleaning by encompassing proactive governance, metadata management, and ongoing monitoring to ensure trustworthiness for analytics.

Q2.  Explain why poor data quality leads to misleading dashboards and incorrect decisions.

-> Poor data quality distorts source data into flawed aggregates, causing dashboards to display incorrect trends or totals. This misguides stakeholders toward flawed strategies, like overstocking based on inflated sales figures.

Q3.  What is duplicate data? Explain three causes in ETL pipelines.

-> Duplicate data means identical or near-identical records appearing multiple times. In ETL, causes include multiple source feeds merging without deduplication, failed idempotent inserts during loads, and transformation errors replicating rows.

Q5. Why should data validation be performed during transformation rather than after loading?

-> Validating during transformation catches errors early, prevents propagating bad data to the target system, and allows efficient rejection or correction before costly reloads.

Q6. Explain how business rules help in validating data accuracy. Give an example.

-> Business rules encode domain logic like valid value ranges or cross-field dependencies to verify accuracy.

Example: In sales data, ensure Txn_Amount > 0 and matches Quantity * unit price within tolerance.
